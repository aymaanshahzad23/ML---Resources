# Cross Validation
<li> Cross Validation is good to diagnose a good model
<li> Divide the dataset into 3, and use 1 for cross validation among the various models

If training error and cross validation error both are large -> data is underfitting (low variance, high bias)
<br>
Solution could be to decrease regularization parameter, to make weights large
<br>
We can also increase features by introducing features like x1*x2 instead of just x1, x2 etc & can make introduce polynomial regression (increase order of polynomial) 

If training error is low and cross validation error is large -> overfitting data (high variance, low bias)
<br>
Solution could be to increase regularization parameter, so as to penalize the weights more and make them small 

---

To increase data to get a better model can be done by Data Augmentation

# Transfer Learning


Train NN on another available large dataset for initial layers suppose whose final layer has a million outputs, 
Then replace the last layer with your desired output but use all/some of the previous trained layers, 

Steps in Transfer learning
1. Supervised Pretraining
2. Fine tuning

Sometimes u just need to do the fine tuning part as the pretrained models are alredy available
